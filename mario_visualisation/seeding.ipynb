```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_B2_13 = pd.read_csv("../results/20240516-19.46.28_B2-long/logs/explain/20240516-19.46.28_B2-long_episodes.csv", header=None)
# add column names
df_B2_13.columns =['timestamp', 'name', 'seed', 'total_steps', 'episode', 'episode_steps', 'reward', 'distance', 'velocity', 'game_time', 'game_score', 'flag', 'loss', 'epsilon', 'dataset']
df_B2_13['ewm_reward'] = df_B2_13['reward'].ewm(alpha=0.01).mean()
df_B2_13['ewm_wins'] = df_B2_13['flag'].ewm(alpha=0.01).mean()
df_B2_13.head(5)
```

```python
df_B2_42 = pd.read_csv("../results/20240602-22.32.40_B2-42/explain/20240602-22.32.40_B2-long_episodes.csv", header=None)
# add column names
df_B2_42.columns =['timestamp', 'name', 'seed', 'total_steps', 'episode', 'episode_steps', 'reward', 'distance', 'velocity', 'game_time', 'game_score', 'flag', 'loss', 'epsilon', 'dataset']
df_B2_42['ewm_reward'] = df_B2_42['reward'].ewm(alpha=0.01).mean()
df_B2_42['ewm_wins'] = df_B2_42['flag'].ewm(alpha=0.01).mean()
df_B2_42.head(5)
```

```python
concatenated = pd.concat([df_B2_13, df_B2_42])
```

```python
sns.set(font_scale=1.1)
sns.set_style("whitegrid")
```

```python
sns.lineplot(x='episode', y='ewm_reward', data=concatenated, hue='seed', palette=sns.color_palette('Set2', n_colors=2)).set(xlabel="Episodes trained", ylabel="EWM rewards (alpha=0.01)")

#plt.title(label="EWM rewards for two differently seeded Slow Decay Baseline agents trained for 15K episodes", loc='center', wrap=True)
plt.savefig("./training_B2_rewards_seeding.png", bbox_inches='tight')
#plt.savefig("./training_B2_rewards_seeding.pdf", format='pdf', bbox_inches='tight')
```

```python
sns.lineplot(x='episode', y='ewm_wins', data=concatenated, hue='seed', palette=sns.color_palette('Set2', n_colors=2)).set(xlabel="Episodes trained", ylabel="EWM wins (alpha=0.01)")

#plt.title(label="EWM wins for two differently seeded Slow Decay Baseline agents trained for 15K episodes", loc='center', wrap=True)
plt.savefig("./training_B2_wins_seeding.png", bbox_inches='tight')
#plt.savefig("./training_B2_wins_seeding.pdf", format='pdf', bbox_inches='tight')
```

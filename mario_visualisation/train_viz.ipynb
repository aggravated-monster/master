```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train_5000eps = pd.read_csv("train_5000eps.csv")
df_train_10000eps = pd.read_csv("train_10000eps.csv")
df_train_10000eps_discriminative = pd.read_csv("train_10000eps_discriminative.csv")
df_train_5000eps.head(5)
```

```python
sns.lineplot(x='episode', y='ewm_reward', data=df_train_5000eps, hue='model_id').set(title="Exponentially weighted mean number of wins over 5K episodes", xlabel="Episodes trained", ylabel="EWM wins (alpha=0.0001)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
sns.lineplot(x='episode', y='ewm_wins', data=df_train_5000eps, hue='model_id').set(title="Exponentially weighted mean wins over 5 episodes", xlabel="Episodes trained", ylabel="EWM wins (alpha=0.0001)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
sns.lineplot(x='episode', y='epsilon', data=df_train_5000eps, hue='model_id').set(title="Epsilon decay over 5K episodes", xlabel="Episodes trained", ylabel="Epsilon (min=0.02)")
```

```python
sns.lineplot(x='episode', y='ewm_reward', data=df_train_10000eps, hue='model_id').set(title="Exponentially weighted mean reward over 10K episodes", xlabel="Episodes trained", ylabel="EWM reward (alpha=0.0001)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

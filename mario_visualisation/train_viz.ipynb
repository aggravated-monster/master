```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("train.csv")
df_train.head(5)
```

```python
sns.lineplot(x='episode', y='ewm_reward', data=df_train, hue='model_id').set(title="Exponentially weighted mean number of wins over 10K episodes", xlabel="Episodes trained", ylabel="EWM wins (alpha=0.0001)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
sns.lineplot(x='episode', y='ewm_wins', data=df_train, hue='model_id').set(title="Exponentially weighted mean reward over 10K episodes", xlabel="Episodes trained", ylabel="EWM reward (alpha=0.0001)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
sns.lineplot(x='episode', y='epsilon', data=df_train, hue='model_id').set(title="Epsilon decay over 10K episodes", xlabel="Episodes trained", ylabel="Epsilon (min=0.02)")
```
